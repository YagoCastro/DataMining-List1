# Questão 1
**O que é uma transformação de atributo? Cite um exemplo de como isto pode ajudar os métodos de aprendizado.**

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
O processo de transfomação dos atributos ou das variáveis é um processo que está ligado a limpeza dos dados dentro de um projeto de ciência de dados. E entende-se como ações que são realizadas nessa etapa:

A preparação de dados é uma etapa crucial no processo de ciência de dados, pois é responsável por organizar, limpar e transformar os dados brutos para que possam ser utilizados na modelagem e avaliação. Essa fase consome as saídas da fase anterior, como os objetivos do negócio e o plano do projeto. O principal objetivo é garantir que os dados estejam consistentes, precisos e adequados para extrair insights e permitir uma boa performance dos modelos. [1]
</div>

<div style="text-align: justify;">
Essa técnica nos permite converter os dados brutos em um formato necessário para que os próximos procedimentos de processamento e modelagem de dados possam ser realizados com eficiência.

Tecnicamente. A transformação de dados altera a estrutura, o formato e o valor dos dados e os torna limpos e utilizáveis ​​para os próximos processos.
</div>

<div style="text-align: justify;">
Pode-se citar as seguites processos que fazem parte da transformação:
- Suavização dos dados: Como por exemplo quando aplicamos em um conjunto de dados desbalanceado um técnica de oversample onde os dados da classe minoritária são adicionados mais vezes até o ponto em que eu tenho um balaceamento igualitário entre as duas classe (considerando uma classe alvo binária), temos nessa situação uma necessidade de suavização por mais que pequena nos dados gerados para garantir um melhor treinamento.
- Dados faltantes: Na própria limpeza dos dados de forma inicial temos que identificar os dados ausentes/faltantes e para esses existe técnicas de transformação para que esse valor saia dessa questão de informação faltante pois é algo ruim para análise e aplicação do modelo, para uma informação 
- Outliers: Também temos o exemplo dos outliers, dados que se distânciam muito 
- Construção de novas variáveis: Pode-se transformar ou agregar a um conjunto de dados um novo atributo a partir de outros atributos existentes na base, por exemplo, se eu tenho a altura e o peso, peço obter o IMC
</div>

# Questão 3
Qual é o motivo pelo qual os dados podem não ser IID (independentes e identicamente distribuídos) no exemplo de filtragem de spam de e-mail da aula?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
Uma das premissas fundamentais do aprendizado de máquina é a suposição de que os dados utilizados para treinar um modelo são independentes e identicamente distribuídos (IID). Essa suposição é fundamental para a aplicação de muitos algoritmos de aprendizado de máquina, pois permite que sejam feitas inferências estatísticas precisas a partir dos dados. Quando dizemos que os dados são independentes, estamos afirmando que as observações não estão relacionadas entre si.
</div>


<div style="text-align: justify;">

- Dependência temporal: Em muitos casos, os dados podem apresentar dependências temporais, ou seja, a ocorrência de uma observação pode influenciar a ocorrência de observações futuras. Isso é comum em séries temporais, onde os dados são coletados ao longo do tempo e podem apresentar padrões temporais. Os padrões nos e-mails podem mudar com o tempo. 
- Contexto específico dos usuários: Os e-mails podem refletir padrões pessoais e comportamentos dos usuários, como o tipo de remetentes, frequência de mensagens e tópicos comuns. 
- Dependências Espaciais: Além disso, os dados podem apresentar dependências espaciais, ou seja, a ocorrência de uma observação pode influenciar a ocorrência de observações próximas no espaço.
</div>  

<a id="Q4"></a>
# Questão 4
Qual é a diferença entre um conjunto de validação e um conjunto de teste?

<font color = "blue"> **Resposta:**</font>


Fonte: [4]

<div style="text-align: justify;">
- Conjunto de Treino: Este conjunto de dados é usado para treinar o modelo. Ele é utilizado para ajustar os parâmetros do modelo. Ele vai ensinar o modelo a reconhecer padrões e relações nos dados.

- Conjunto de Validação: Usado durante o treinamento para ajustar os hiperparâmetros (como a taxa de aprendizado, número de camadas, número de árvores em um modelo de ensemble, etc.). **Isso acontece porque é mais seguro para o projeto realizar um validação/ajuste com dados diferentes daqueles que foram utilizados para moldar/treinar o modelo.** Permite que o modelo generalize bem para novos dados, evitando problemas como overfitting (modelo ajustado demais aos dados de treino) ou underfitting (modelo subajustado).
    
    -- Após cada iteração de treinamento, o modelo é avaliado com os dados de validação.
    
    -- A métrica de desempenho (por exemplo, acurácia, F1-score, etc.) informa se o modelo está aprendendo ou se está começando a superajustar (overfitting).

- Conjunto de Teste: Usado apenas uma vez no final do processo de treinamento e validação para avaliar o desempenho real do modelo **em dados nunca vistos antes.** Medir a capacidade de generalização do modelo em dados completamente novos e fornecer uma estimativa final de sua performance.
    
</div> 

<div style="text-align: justify;">
Resumo:
- Dados de treino são vistos pelo modelo e usados para ajustar os parâmetros.
- Dados de validação ajudam a evitar o overfitting ajustando os hiperparâmetros.
- Dados de teste avaliam a generalização e garantem que o modelo funcione bem em dados futuros.
</div>    

<a id="Q5"></a>
# Questão 5
Por que não podemos (normalmente) usar o erro de treinamento para selecionar um hiperparâmetro?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
O ajuste de hiperparâmetros seleciona hiperparâmetros que darão bom desempenho de generalização. Normalmente, isso funciona estimando o desempenho de generalização para diferentes escolhas de hiperparâmetros (por exemplo, usando um conjunto de validação) e, em seguida, escolhendo o melhor. Mas, como acima, essa estimativa será superotimista se os mesmos dados tiverem sido usados ​​para treinamento. A consequência é que hiperparâmetros subótimos serão escolhidos. Em particular, haverá um viés em direção a modelos de alta capacidade que serão superajustados.[5]

Logo, Não podemos usar o erro de treinamento para selecionar hiperparâmetros porque ele não reflete a capacidade do modelo de generalizar para novos dados. O erro de treinamento mede apenas quão bem o modelo se ajusta aos dados nos quais foi treinado, mas não informa sobre sua performance em dados que nunca viu antes. 

O objetivo do ajuste de hiperparâmetros é equilibrar a troca de viés-variância. O viés é a divergência entre as previsões de um modelo e a realidade. Modelos que são subajustados, ou mal ajustados, não conseguem discernir relações fundamentais entre pontos de dados e são incapazes de tirar as conclusões exigidas para um desempenho preciso. A variância é a sensibilidade de um modelo a novos dados. Um modelo confiável deve fornecer resultados consistentes ao migrar de seus dados de treinamento para outros conjuntos de dados. No entanto, modelos com altos níveis de variância são complexos demais: eles são superajustados aos seus conjuntos de dados de treinamento originais e têm dificuldade para acomodar novos dados. [6] 

Modelos com baixo viés são precisos, enquanto modelos com baixa variância são consistentes. Um bom ajuste de hiperparâmetros otimiza para criar o melhor modelo para o trabalho e, ao mesmo tempo, maximizar a eficiência dos recursos computacionais durante o treinamento. [6]
</div>

# Questão 6
Naïve Bayes faz a suposição de que todos os recursos são condicionalmente independentes, dado o rótulo da classe. Por que essa suposição é necessária e o que aconteceria sem ela?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
O algoritmo Naïve Bayes faz uma suposição bastante ingênua, mas útil para simplificar o cálculo de probabilidades. Ele assume que cada atributo de uma instância (por exemplo, características de um objeto) influencia independentemente a classificação dessa instância em uma determinada classe (por exemplo, categorias de objetos).

A suposição de independência condicional em Naïve Bayes é necessária para simplificar os cálculos, tornando o modelo escalável e eficiente. Sem essa suposição, o modelo exigiria mais dados, maior poder computacional e seria mais complexo de treinar e interpretar. 
</div>     

<a id="Q7"></a>
# Questão 7
Por que o KNN é considerado um método não paramétrico e quais são as consequências indesejáveis do design não paramétrico do KNN?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
O algoritmo k-nearest neighbors (KNN), ou k vizinhos mais próximos, é um classificador supervisionado não paramétrico, que usa proximidade para fazer classificações ou previsões sobre o agrupamento de um ponto de dados individual. Ele não assume nenhuma hipótese explícita sobre a distribuição dos dados subjacentes. Isso significa que o modelo não realiza nenhum treinamento propriamente dito para ajustar parâmetros antes de fazer previsões. Em vez disso, ele armazena os dados de treinamento e utiliza a proximidade entre os pontos para fazer classificações ou regressões.

Não escala bem: como o KNN é um algoritmo preguiçoso, ele consome mais memória e armazenamento de dados em comparação com outros classificadores. Isso pode ser caro do ponto de vista de tempo e dinheiro. 

Sensibilidade ao ruído:O KNN pode ser influenciado negativamente por outliers ou ruído nos dados de treinamento, já que os pontos mais próximos são usados diretamente para classificar ou prever.

Maldição da dimensionalidade: Em espaços de alta dimensionalidade, as distâncias entre os pontos tornam-se menos significativas, e a performance do KNN pode ser drasticamente afetada. Isso ocorre porque os pontos começam a parecer igualmente distantes uns dos outros. Isso às vezes também é chamado de fenômeno de pico, onde, após o algoritmo atingir o número ideal de funcionalidades, a adição de mais delas aumenta a quantidade de erros de classificação, especialmente quando o tamanho da amostra é menor.
</div>

# Questão 8
Para qualquer modelo paramétrico, como o aumento do número de exemplos de treinamento afeta as duas partes do custo-benefício fundamental?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
- O aumento do número de exemplos de treinamento reduz a variância, pois o modelo tem mais dados para aprender a verdadeira distribuição, tornando-o menos dependente de ruídos ou peculiaridades do conjunto de treinamento.

- O aumento do número de exemplos de treinamento geralmente não tem impacto significativo no viés, pois o viés depende principalmente da escolha da arquitetura ou do tipo de modelo. Por exemplo, se o modelo é linear, ele não conseguirá capturar relações não lineares, independentemente de quantos dados sejam adicionados.
</div> 

<div style="text-align: justify;">
Resumo:
- O aumento do número de exemplos de treinamento reduz a variância, levando a um modelo mais robusto e menos propenso a overfitting.
- No entanto, o aumento dos dados não reduz o viés, a menos que a complexidade do modelo seja ajustada para melhorar sua capacidade de capturar a verdadeira relação nos dados.
</div>

# Questão 9
Tanto o aprendizado supervisionado quanto os modelos de agrupamento recebem uma entrada x_i e produzem um rótulo y_i. Qual é a principal diferença?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
- A principal diferença entre as duas abordagens é a classificação entre conjuntos de dados rotulados e não rotulados. O aprendizado supervisionado utiliza conjuntos de dados com rótulos para treinar algoritmos de classificação ou previsão. Dados identificados como "treinamento" são adicionados, e o modelo ajusta de forma iterativa a maneira como avalia diferentes características até alcançar o resultado desejado. Modelos de aprendizado supervisionado são mais rigorosos em comparação ao método de aprendizagem sem supervisão. Contudo, demandam o envolvimento do usuário durante o processamento de dados para garantir que a rotulagem seja feita adequadamente. [8]

- Por outro lado, modelos de aprendizado não supervisionado sempre operam sem a intervenção humana. Eles detectam e atingem uma estrutura composta por dados não rotulados. A única colaboração do usuário aqui é a validação das variáveis de saída. Por exemplo, quando alguém compra um novo laptop on-line, um modelo de aprendizado não supervisionado entende que o usuário pertence a um grupo de clientes que compram uma série de produtos relacionados em conjunto. No entanto, a função do analista de dados é garantir que o sistema de recomendação possa oferecer opções para bolsas de laptop, protetores de tela e carregadores de carro. [8]
</div>

# Questão 10
No agrupamento k-means, os clusters são garantidamente regiões convexas. As áreas que recebem o mesmo rótulo por KNN também são convexas?

<font color = "blue"> **Resposta:**</font>

<div style="text-align: justify;">
- K-Means:
No K-means, os clusters são formados minimizando a soma das distâncias quadradas entre os pontos e o centróide de cada cluster. Essa abordagem leva a regiões convexas, pois a minimização da distância euclidiana resulta em divisores lineares (hiperplanos) entre os clusters.
- K-Nearest Neighbors (KNN):
No KNN, um ponto recebe o rótulo com base nos K vizinhos mais próximos no espaço. As regiões associadas a um rótulo dependem da distribuição dos pontos no conjunto de treinamento.
As fronteiras entre os rótulos não são lineares nem necessariamente simples. Elas podem ter formas complexas e não convexas, já que a decisão é baseada em uma votação local dos vizinhos mais próximos, que depende da posição dos pontos no conjunto de treinamento. As regiões podem ter formas muito complexas, devido à influência local de cada ponto de treinamento.
