## 1.1.Construction of the Data Dictionary (Descrição dos Dados)


Nessa primeira etapa é muito importante mapeamos os nossos dados, primeiro para conhecer cada um e identificar qual característica ou informação aquele dado expressa. Documentação e Auditoria: Em projetos de longo prazo, o dicionário de dados serve como um registro que documenta todas as variáveis e suas modificações. Isso é importante para auditorias, para rastrear a origem e a finalidade dos dados e entender mudanças feitas ao longo do tempo.

# 4.Prediction_Model


## 4.1.First_Model

```python
df4_1 = df.copy()
```

Esse primeiro modelo será criado como base, ele será criado desconsiderando as colunas do tipo qualitativa, para temos um valor como base de forma inicial.

```python
# Criando uma lista com as colunas numéricas
col_num = df4_1.select_dtypes(include=['number']).columns.tolist()
```

```python
# Eliminando as colunas do tipo object:
print(f'Colunas presentes no dataset antes: {df4_1.info()}')
df4_1 = df4_1.loc[:, col_num]
print(f'Colunas presentes no dataset depois: {df4_1.info()}')
```

```python
# Selecionando X e y:
X = df4_1.drop(columns='charges', axis=1)
y = df4_1.charges
```

```python
# Separando a base em treino e teste:
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=42)
```

```python
# Dicionários para armazenar métricas e informações dos modelos
modelos_erro_abs           = {}  
modelos_erro_square        = {}  
```

```python
np.random.seed(42)

modelos = {
    "LinearRegression": LinearRegression(),
    "Tree": tree.DecisionTreeRegressor(random_state=42),
    "KNN": KNeighborsRegressor(n_neighbors=250)
}
```

```python
# Função para Treinamento dos Modelos e Obtenção das métricas para cada
# modelo quando aplicado aos dados de teste
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
       Função com o objetivo de Treinar e Testar determinados modelos de Machine Learning
       modelos: Um dicionário de diferentes modelos de aprendizado de máquina do scikit-learn
       X_train: Dados de treinamento
       X_test: Dados de teste
       y_train: Valores reais de treino
       y_test: Valores reais de teste
    """
    # Dicionários para armazenar os erros
    modelos_erro_abs = {}
    modelos_erro_square = {}

    # Loop para o dicionário dos modelos
    for name, model in modelos.items():
        # Ajustando o modelo aos dados
        model.fit(X_train, y_train)
        
        # Calculando o Erro Médio Absoluto
        modelos_erro_abs[name] = mean_absolute_error(y_test, model.predict(X_test))
        
        # Calculando o Erro Quadrático Médio
        modelos_erro_square[name] = mean_squared_error(y_test, model.predict(X_test))

    # Gerar gráficos para cada modelo
    fig, ax = plt.subplots(ncols=len(modelos), figsize=(18, 5))
    for i, (name, model) in enumerate(modelos.items()):
        ax[i].scatter(y_test, model.predict(X_test))
        ax[i].plot([0, 70000], [0, 70000], '--r')
        ax[i].set_title(name)
        ax[i].set_xlabel('Valores reais')
        ax[i].set_ylabel('Previsões')
        ax[i].set(xlim=(0, 70000), ylim=(0, 70000))

    plt.show()

    # Exibir as métricas de erro
    print(f"====Modelos====\n")

    df_resultados = pd.DataFrame({
        "Erro_Absoluto": modelos_erro_abs,
        "Erro_Quadrático": modelos_erro_square,
    })
    print(df_resultados)
```

```python
start_time = time.time()
#Chamando a função
model_scores = fit_and_score(
                             models = modelos,
                             X_train = X_train,
                             X_test = X_test,
                             y_train = y_train, 
                             y_test = y_test)
model_scores

end_time =time.time()
print(f"Tempo de execução: {end_time - start_time:.4f} segundos")
```
